```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
top_size , bottom_size , left_size,right_size = (50,50,50,50)
img = cv2.imread("img/000064.jpg")
img = cv2.cvtColor(img,cv2.COLOR_RGB2BGR)
```

```python
i1 =  cv2.copyMakeBorder(img,top_size,bottom_size,left_size,right_size,borderType=cv2.BORDER_REPLICATE)
i2 =  cv2.copyMakeBorder(img,top_size,bottom_size,left_size,right_size,borderType=cv2.BORDER_REFLECT)
i3 =  img+20
i4 =  cv2.copyMakeBorder(img,top_size,bottom_size,left_size,right_size,borderType=cv2.BORDER_CONSTANT,value=(128,128,128))
```

```python
plt.subplot(231)
plt.imshow(img,'gray')
plt.subplot(232)
plt.imshow(i1,'gray')
plt.subplot(233)
plt.imshow(i2,'gray')
plt.subplot(234)
plt.imshow(i3,'gray')
plt.subplot(235)
plt.imshow(i4,'gray')
type(img)
```

```python
def show(img):
    cv2.imshow("img",img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
show(img)
```

```python
cv2.rectangle(img,(0,0),(200,200),(0,255,0),3)
show(img)
```

```python
font = cv2.FONT_HERSHEY_PLAIN
cv2.putText(img,"OPENCV",(10,500),font,1,(255,255,255),1,cv2.LINE_AA)
show(img)
```
